```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import yfinance as yf
```

[PT] Definindo a data de analise

```python
inicio = '2015-01-01'
fim = ''
```

[PT] Realizando o download do arquivo

```python
petr = yf.download('PETR4.SA', start=inicio, end_date=fim, interval='1d')
```

```python
petr.tail()
```

[PT] Obtendo o fechamento em porcentagem e criando uma coluna de abertura do dia seguinte em porcentagem

```python
petr['Fechamento em %']= petr['Close'].pct_change()*100
```

```python
petr['Abertura do dia seguinte %'] = ((petr['Open'].shift(-1)-petr['Close'])/petr['Close'])*100
```

```python
petr
```

[PT] Criando o filtro de analise de cotações

```python
f = -10
filtro = petr.loc[petr['Fechamento em %']<=(f)]
```

[PT] Excluindo algumas colunas para facilitar a visualização

```python
filtro.drop(['High'], axis=1, inplace=True)
filtro.drop(['Low'], axis=1, inplace=True)
filtro.drop(['Volume'], axis=1, inplace=True)
filtro.drop(['Adj Close'], axis=1, inplace=True)
```

```python
filtro
```

[PT] Calculando a probabilidade de abrir em alta no dia seguinte.

```python
resultado = filtro['Abertura do dia seguinte %']
Resultado_bin = []

for i in resultado:
  if i > 0:
    Resultado_bin.append('Positivo')
  else:
    Resultado_bin.append('Negativo')

filtro['Resultado'] = Resultado_bin
```

```python
c = 0
for i in filtro['Resultado']:
  if i =='Positivo':
    c=c+1
prob = (c/len(filtro['Resultado']))*100
print(f'O ativo tem uma probabilidade de {prob}% de abrir positivo quando fecha abaixo de {f}% desde {inicio} até hoje.')
```
